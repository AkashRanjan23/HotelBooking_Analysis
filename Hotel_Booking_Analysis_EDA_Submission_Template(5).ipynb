### **Step 1: Handling missing values**

````python
# Write your code to make your dataset analysis ready.
hotel_data.isnull().sum().sort_values(ascending=False)[:5]
````

**Earlier we had already mentioned that there was only Four columns that were having some Missing values in them and that was :(i).children,    (ii).agent,
                                               (iii).countries,   (iv).company.
                                               Now Let's replace those missing values in the above mentioned column with some meaningful value**


````python
# Replace missing values in the specified columns with some meaningful Value:
hotel_data['company'].fillna(0,inplace=True)
hotel_data['agent'].fillna(0,inplace=True)
hotel_data['children'].fillna(0,inplace=True)
hotel_data['country'].fillna('Unknown',inplace=True)
````

### What all manipulations have you done and insights you found?

**In the 'company' and 'agent' columns of the dataset, the entries correspond to company numbers and agent numbers. Nevertheless, there are occurrences where customers made direct hotel bookings without the involvement of any agent or booking company. To address this, instances where no agent or company is involved will be substituted with the value '0'. Similarly, in the 'children' column, missing data entries will be replaced with '0' to account for cases where the information was not provided. Additionally, in the 'country' column, instances with missing values were addressed by replacing them with 'Unknown'.**

````python
# Now there is no missing data in my DataFrame as all the missing values have been replaced with appropriate values, ensuring a more comprehensive
# and reliable dataset for further analysis and interpretation.
hotel_data.isnull().sum().sort_values(ascending=False)[:5]
````

### **Step 2: Removing duplicates**

````python
# @title Getting The Total count of all the Duplicate Rows in the Hotel_data DataFrame, Where "False" represents the Total number of count of "Non-Duplicate" Values and "True" represents the Total number of count of "Duplicate" Values:

Duplicate_Rows=hotel_data.duplicated().value_counts()
print(Duplicate_Rows)
````

````python
# @title "Now, let's represent the 'Duplicate' and 'Non-Duplicate' rows using a "Bar Graph" Visualization."

plt.figure(figsize=(10,6))
sns.barplot(x=Duplicate_Rows.index, y=Duplicate_Rows.values, palette="viridis")
plt.title("Representation of 'Duplicate' & 'Non-Duplicate' Rows in DataFrame")
plt.xlabel('Duplicate Rows')
plt.ylabel('Count')

plt.show()
````

````python
# @title Let's enhance the clarity of our Hotel_data DataFrame by removing all duplicate rows, thereby streamlining our dataset.:

hotel_data.drop_duplicates(inplace=True)
````

````python
# @title Let's finally check the shape of our Hotel_data DataFrame after filling all the "Null_Values" and removing all the "Duplicate_Rows".

print(hotel_data.shape)
print("Number of Rows after droping all the duplicate Rows:",hotel_data.shape[0])
print("Number of Columns:",hotel_data.shape[1])
````

### **Step 3: Converting columns to proper dtype and format.**

````python
# Converting the data type of specific columns (children, agent, company) from 'float64' to 'int64':

hotel_data[['children','company','agent']] = hotel_data[['children','company','agent']].astype('int64')

# Let's convert the column name 'reservation_status_date' from 'Object' datatype to 'datetime':

hotel_data['reservation_status_date'] = pd.to_datetime(hotel_data['reservation_status_date'], format='%Y-%m-%d')
````

### **Step 4: Adding some columns.**

````python
# Adding a 'total_stay' column in the DataFrame, calculated as the sum of 'stays_in_week_nights' and 'stays_in_weekend_nights':
hotel_data['Total_Stay']=hotel_data['stays_in_week_nights']+hotel_data['stays_in_weekend_nights']

# Adding a 'total_people' column in the DataFrame, calculated as the sum of 'adults', 'children', and 'babies'
hotel_data['total_people'] = hotel_data['adults'] + hotel_data['children'] + hotel_data['babies']
````

#**"Most Bookings Achieved: Top 5 Agents Highlighted"**

````python
# Chart - 10 visualization code
# Creating a dataset by grouping by agent column and it's count
# Just taken all the Top 5 Agent having more Booking_Count
agent_booking_count = hotel_data['agent'].value_counts().head(10).reset_index().rename(columns={'index': 'Agent_Code', 'agent': 'Total_Booking_Count'}).sort_values(by='Total_Booking_Count', ascending=False)

# Extracting top 5 agents by booking count
top_agent = agent_booking_count[:5]

# Plotting the values
fig, ax = plt.subplots(figsize=(15, 7))
wedge_properties = {'linewidth': 1, 'edgecolor': "#39FF14"}
explode = (0.01, 0.01, 0.01, 0.01, 0.01)

# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100. * np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)

# Pie chart
wedges, texts, autotexts = ax.pie(top_agent['Total_Booking_Count'],
                                  autopct=lambda pct: func(pct, top_agent['Total_Booking_Count']),
                                  explode=explode,
                                  shadow=False,
                                  startangle=50,
                                  wedgeprops=wedge_properties)

# Adding legend
ax.legend(wedges, top_agent['Agent_Code'],
          title="agents",
          loc="upper left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
ax.set_title("The top five agents based on Booking", fontsize=20)

# Show chart
plt.axis('equal')
plt.show()
````

##### 1. Why did you pick the specific chart?

**Answer Here:**


A pie chart is chosen for the top 5 agents by booking due to its ability to visually represent the proportional contribution of each agent to the total bookings. The chart provides a clear and concise overview of the distribution, with percentages displayed, making it easy to identify and compare the booking volumes of the top agents at a glance.
